```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definindo a função f(x, y) = sin(x) * cos(y)
def f(x, y):
    return np.sin(x) * np.cos(y)

# Criando a grade de pontos
x = np.linspace(-3, 3, 50)
y = np.linspace(-3, 3, 50)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Gráficos 3D
fig = plt.figure(figsize=(12, 4))

# Gráfico 3D
ax1 = fig.add_subplot(131, projection='3d')
ax1.plot_surface(X, Y, Z, cmap='viridis')
ax1.set_title('Gráfico 3D')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('z')

# Curvas de Nível
ax2 = fig.add_subplot(132)
ax2.contour(X, Y, Z, levels=20, cmap='plasma')
ax2.set_title('Curvas de Nível')
ax2.set_xlabel('x')
ax2.set_ylabel('y')

# Mapa de Contorno
ax3 = fig.add_subplot(133)
contour = ax3.contourf(X, Y, Z, levels=20, cmap='coolwarm')
plt.colorbar(contour, ax=ax3)
ax3.set_title('Mapa de Contorno')
ax3.set_xlabel('x')
ax3.set_ylabel('y')

# Exibir gráficos
plt.tight_layout()
plt.show()
```
